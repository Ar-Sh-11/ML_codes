```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the training datasets
train=pd.read_csv('Amex_Campus_Challenge_Train.csv')
```

```python
#these have more than 90% rows with null values hence needed to be dropped off
drop_col=['customer_merchant_02',
'customer_merchant_01',
'customer_digital_activity_07',
'customer_digital_activity_08',
'customer_digital_activity_09',
'customer_digital_activity_18',
'merchant_spend_11',
'customer_digital_activity_04']
train_ad=train.drop(columns=drop_col)
```

```python
pred=train_ad[['ind_recommended','activation']]
feat=train_ad.drop(columns=['activation','ind_recommended'])
```

```python
featwna=feat.fillna(feat.mean())
```

```python
pred.isna().sum()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(featwna, pred, test_size=0.1, random_state=42)
```

```python
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
# Make predictions
y_pred = classifier.predict(X_test)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(f"MSE: {mse:.2f}")
```

```python
import joblib
joblib.dump(classifier,'xgboost_classifier2(0.9,0.1,42)_model.joblib')
```
